```python
from mpl_toolkits import mplot3d
import numpy
import matplotlib.pyplot as plt
import math

def undirectional_component(boundary, n):
    values = []
    for i in range(0,360):
        sine_component = math.sin((n*math.pi*i)/boundary)
#         print(math.sqrt(2/boundary)*sine_component)
        values.append(math.sqrt(2/boundary)*sine_component)
    return values

x = numpy.array(undirectional_component(300,25))
y = numpy.array(undirectional_component(25,300))
ax = plt.axes(projection='3d')



z = [x[i]*y[i] for i in range(len(x))]
# print(z)

ax.plot_trisurf(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title("Wave function without probability")



```

```python

def undirectional_proability_component(boundary, n):
    values = []
    for i in range(0,360):
        sine_component = math.sin((n*math.pi*i)/boundary)
#         print(math.sqrt(2/boundary)*sine_component)
        values.append((math.sqrt(2/boundary)*sine_component)**2)
    return values



x = numpy.array(undirectional_proability_component(300,25))
y = numpy.array(undirectional_proability_component(2,25))

ax = plt.axes(projection='3d')

z = [x[i]*y[i] for i in range(len(x))]

ax.plot_trisurf(x, y, z,cmap='plasma', edgecolor='none')
ax.set_title("Wave function with probability")
```
